```python
from pandas import read_excel 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from pylab import rcParams

rcParams['figure.figsize'] = 12, 8 #Sets a reasonable figure size for this notebook
```

```python

#extra wide range of binding energies data is within
BEmin = 270
BEmax = 305
alldata = pd.DataFrame(index=(np.arange(BEmin,BEmax,0.1).round(1))) 

locations=[  #Structured as [.xlsx datafile, sheet number (or name), columns to pull from, rows to skip]
    [file1, 1, 'B,C', 3], 
    [file1, 2, 'B,C', 3], 
    [file1, 3, 'B,C', 3], 
    [file1, 4, 'B,C', 3], 
]

for num, param in enumerate(locations):
    data = read_excel(param[0],
                     sheet_name=param[1],
                     names = ['BE','CPS'+str(num)],
                     usecols = param[2],
                     skiprows=param[3])
    data = data.dropna()
    data['CPS'+str(num)] -= (data['CPS'+str(num)].min())
    plt.plot(data['BE'],data['CPS'+str(num)], color='grey', linewidth=1)
    
    data2 = data.set_index(round(data['BE'],1))
    data2 = data2.drop('BE',axis=1)
    alldata = pd.concat([alldata,data2], axis=1, join='inner')

alldata['avg'] = alldata.mean(axis=1)
plt.plot(alldata.index.values,alldata['avg'], color='black', linewidth=2, label="Slurry")

alldata1 = pd.DataFrame(index=(np.arange(BEmin,BEmax,0.1).round(1))) 

locations1=[  #Structured as [.xlsx datafile, sheet number (or name), columns to pull from, rows to skip]
    [file1, 1, 'B,D', 3], 
]

for num, param in enumerate(locations1):
    data1 = read_excel(param[0],
                     sheet_name=param[1],
                     names = ['BE','CPS'+str(num)],
                     usecols = param[2],
                     skiprows=param[3])
    data1 = data1.dropna()
    data1['CPS'+str(num)] -= (data1['CPS'+str(num)].min())
    plt.plot(data1['BE'],data1['CPS'+str(num)], color='red', linewidth=1, label='Adventitious Carbon')
    
    data3 = data1.set_index(round(data1['BE'],1))
    data3 = data3.drop('BE',axis=1)
    alldata1 = pd.concat([alldata1,data3], axis=1, join='inner')
    
alldata2 = pd.DataFrame(index=(np.arange(BEmin,BEmax,0.1).round(1)))   

locations2=[  #Structured as [.xlsx datafile, sheet number (or name), columns to pull from, rows to skip]
    [file1, 1, 'B,E', 3], 
]

for num, param in enumerate(locations2):
    data2 = read_excel(param[0],
                     sheet_name=param[1],
                     names = ['BE','CPS'+str(num)],
                     usecols = param[2],
                     skiprows=param[3])
    data2 = data2.dropna()
    data2['CPS'+str(num)] -= (data2['CPS'+str(num)].min())
    plt.plot(data2['BE'],data2['CPS'+str(num)], color='blue', linewidth=1, label='Super P')
    
    data4 = data2.set_index(round(data2['BE'],1))
    data4 = data4.drop('BE',axis=1)
    alldata2 = pd.concat([alldata2,data4], axis=1, join='inner')
    
alldata3 = pd.DataFrame(index=(np.arange(BEmin,BEmax,0.1).round(1)))     

locations3=[  #Structured as [.xlsx datafile, sheet number (or name), columns to pull from, rows to skip]
    [file1, 1, 'B,F', 3], 
]

for num, param in enumerate(locations3):
    data3 = read_excel(param[0],
                     sheet_name=param[1],
                     names = ['BE','CPS'+str(num)],
                     usecols = param[2],
                     skiprows=param[3])
    data3 = data3.dropna()
    data3['CPS'+str(num)] -= (data3['CPS'+str(num)].min())
    plt.plot(data3['BE'],data3['CPS'+str(num)], color='green', linewidth=1, label='Singly-oxygenated Carbon')
    
    data5 = data3.set_index(round(data3['BE'],1))
    data5 = data5.drop('BE',axis=1)
    alldata3 = pd.concat([alldata3,data5], axis=1, join='inner')
    
alldata4 = pd.DataFrame(index=(np.arange(BEmin,BEmax,0.1).round(1)))     

locations4=[  #Structured as [.xlsx datafile, sheet number (or name), columns to pull from, rows to skip]
    [file1, 1, 'B,G', 3], 
]

for num, param in enumerate(locations4):
    data4 = read_excel(param[0],
                     sheet_name=param[1],
                     names = ['BE','CPS'+str(num)],
                     usecols = param[2],
                     skiprows=param[3])
    data4 = data4.dropna()
    data4['CPS'+str(num)] -= (data4['CPS'+str(num)].min())
    plt.plot(data4['BE'],data4['CPS'+str(num)], color='orange', linewidth=1, label='Carboxyl Carbon')
    
    data6 = data4.set_index(round(data4['BE'],1))
    data6 = data6.drop('BE',axis=1)
    alldata4 = pd.concat([alldata4,data6], axis=1, join='inner')
    
alldata5 = pd.DataFrame(index=(np.arange(BEmin,BEmax,0.1).round(1)))     

locations5=[  #Structured as [.xlsx datafile, sheet number (or name), columns to pull from, rows to skip]
    [file1, 1, 'B,I', 3], 
]

for num, param in enumerate(locations5):
    data5 = read_excel(param[0],
                     sheet_name=param[1],
                     names = ['BE','CPS'+str(num)],
                     usecols = param[2],
                     skiprows=param[3])
    data5 = data5.dropna()
    data5['CPS'+str(num)] -= (data5['CPS'+str(num)].min())
    plt.plot(data5['BE'],data5['CPS'+str(num)], color='magenta', linewidth=1, label='Fit')
    
    data7 = data5.set_index(round(data5['BE'],1))
    data7 = data7.drop('BE',axis=1)
    alldata5 = pd.concat([alldata5,data7], axis=1, join='inner')
    

plt.xlabel('Binding Energy (eV)',fontsize=22)
plt.ylabel('Subtracted Counts (a.u.)',fontsize=22)
plt.xlim(293,280)
plt.ylim(0,6000)
plt.xticks(fontsize=18, rotation=0)
plt.yticks(fontsize=18, rotation=0)
plt.legend(fontsize=18)
plt.show()
```
